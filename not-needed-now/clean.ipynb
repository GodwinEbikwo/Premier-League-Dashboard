```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


misc_stats = pd.read_csv("squad_misc_stats.csv")
df = pd.read_csv("./src/data/scrapped_player_stats-18th-Aug.csv")
```

```python
df.info()
```

```python
df['nation'] = df['nation'].str.extract(r'([A-Z]{3})')
```

```python
# List of columns with float64 dtype
float64_columns = ['age', 'min', '90s', 'gls', 'ast', 'g+a', 'g-pk', 'pk', 'pkatt', 'crdy', 'crdr', 'xg', 'npxg', 'xag', 'npxg+xag']

# Convert selected columns to float32 dtype
df[float64_columns] = df[float64_columns].fillna(0).astype('int')
```

```python
df.info()
df = df.rename(columns=lambda x: x.lower())
```
